## 1. What is Air Pollution?

Air pollution refers to the presence of harmful substances in the atmosphere, including gases and particulate matter, that negatively impact human health and the environment.

### Common Sources

* Vehicle emissions
* Industrial and factory pollutants
* Dust, pollen, and biological particles

In urban areas, **ozone (O₃)** plays a critical role in forming **smog**, a visible form of air pollution.

---

## 2. Health Impact

Exposure to air pollution can lead to:

* Respiratory diseases (e.g., asthma)
* Cardiovascular complications
* Long-term lung damage

### High-Risk Groups

* Children
* Elderly individuals
* People with pre-existing heart or lung conditions

> Air pollution exists both **outdoors and indoors**, impacting overall health.

---

## 3. Major Air Pollutants

### 3.1 Particulate Matter (PM)

A mixture of solid and liquid particles suspended in air:

* Dust, soot, smoke
* Sulphates, nitrates
* Organic chemicals and metals

#### Types by Size

* **PM10** ≤ 10 microns
* **PM2.5** ≤ 2.5 microns (most harmful)
* **Ultrafine** ≤ 0.1 microns

**Insight:** Smaller particles penetrate deeper into lungs → higher health risk

---

### 3.2 Nitrogen Dioxide (NO₂)

* Toxic gas from:

  * Vehicles (especially diesel)
  * Power plants
  * Industrial processes

**Key Characteristics:**

* Higher near traffic zones
* Major contributor to urban pollution

---

### 3.3 Ozone (O₃)

* Gas composed of three oxygen atoms

#### Dual Role

* Upper atmosphere → protects from UV radiation
* Ground level → harmful pollutant

#### Formation

* Sunlight + Nitrogen Oxides (NOx) + VOCs

#### Behavior

* Peaks in summer and afternoons
* Often higher in rural regions

**Role:** Major contributor to smog

---

### 3.4 Sulphur Dioxide (SO₂)

* Colorless gas with pungent smell

**Sources:**

* Burning fossil fuels (coal, oil)
* Power plants
* Industrial activities

**Impact:**

* Travels long distances
* Contributes to secondary pollutants

```python
def missing_values_table(df):
        # Total missing values
        mis_val = df.isnull().sum()
        
        # Percentage of missing values
        mis_val_percent = 100 * df.isnull().sum() / len(df)
        
        # Make a table with the results
        mis_val_table = pd.concat([mis_val, mis_val_percent], axis=1)
        
        # Rename the columns
        mis_val_table_ren_columns = mis_val_table.rename(
        columns = {0 : 'Missing Values', 1 : '% of Total Values'})
        
        # Sort the table by percentage of missing descending
        mis_val_table_ren_columns = mis_val_table_ren_columns[
            mis_val_table_ren_columns.iloc[:,1] != 0].sort_values(
        '% of Total Values', ascending=False).round(1)
        
        # Print some summary information
        print ("Your selected dataframe has " + str(df.shape[1]) + " columns.\n"      
            "There are " + str(mis_val_table_ren_columns.shape[0]) +
              " columns that have missing values.")
        
        # Return the dataframe with missing information
        return mis_val_table_ren_columns
```

```python
#Imputing AQI_bucket(Categorical) missing values according to aqi

def get_aqi_bucket(aqi):
    if pd.isna(aqi):
        return None
    elif aqi <= 50:
        return "Good"
    elif aqi <= 100:
        return "Satisfactory"
    elif aqi <= 200:
        return "Moderate"
    elif aqi <= 300:
        return "Poor"
    elif aqi <= 400:
        return "Very Poor"
    else:
        return "Severe"
    

    
```

# Monthly Trend Analysis — Delhi AQI Dataset

## Objective

Visualize long-term pollution trends by aggregating daily data into monthly averages.

---

## Why Monthly Aggregation?

Daily data is:

* Noisy
* Overcrowded on plots
* Hard to interpret

Monthly aggregation:

* Smooths fluctuations
* Highlights trends
* Reveals seasonality

---

## Approach

1. Convert Date column to datetime
2. Extract Year-Month
3. Aggregate pollutant values (mean)
4. Plot using clean time axis

---


```python
# Ensure datetime
delhi_data['Date'] = pd.to_datetime(delhi_data['Date'])

# Sort
delhi_data = delhi_data.sort_values('Date')

# Create Year-Month
delhi_data['YearMonth'] = delhi_data['Date'].dt.to_period('M')

# Aggregate monthly mean
monthly_data = delhi_data.groupby('YearMonth').mean(numeric_only=True).reset_index()

# Convert to timestamp
monthly_data['YearMonth'] = monthly_data['YearMonth'].dt.to_timestamp()
```

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

pollutants = ['PM2.5', 'PM10', 'NO', 'NO2', 'NOx', 'NH3', 'CO', 'SO2', 'O3', 'AQI']

for pollutant in pollutants:
    plt.figure(figsize=(12,6))
    
    plt.plot(monthly_data['YearMonth'], monthly_data[pollutant])
    
    # Highlight lockdown
    plt.axvspan(pd.to_datetime('2020-03-01'),
                pd.to_datetime('2020-07-01'),
                color='red', alpha=0.2, label='Lockdown')
    
    # Format x-axis
    plt.gca().xaxis.set_major_locator(mdates.YearLocator())
    plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
    
    plt.title(f"{pollutant} Trend (2015–2020)")
    plt.xlabel("Year")
    plt.ylabel(pollutant)
    
    plt.legend()
    plt.tight_layout()
    
    plt.show()
```

## Lockdown Impact — Quantitative Analysis

This table compares average pollutant levels between:

* Pre-lockdown period (2019)

---

### Metrics Used

* **Absolute Change** → direct difference in pollutant levels
* **Percentage Change (%)** → relative improvement or deterioration

---

### Interpretation

* Negative % change → Improvement (pollution reduced)
* Positive % change → Deterioration (pollution increased)

---

### Key Insight

This analysis quantifies the impact of reduced human activity during lockdown on air quality.

---


```python
import pandas as pd
import numpy as np

# Ensure datetime
delhi_data['Date'] = pd.to_datetime(delhi_data['Date'])

# Define precise windows
pre_lockdown = delhi_data[
    (delhi_data['Date'] >= '2019-03-24') &
    (delhi_data['Date'] <= '2019-05-31')
].copy()

lockdown = delhi_data[
    (delhi_data['Date'] >= '2020-03-24') &
    (delhi_data['Date'] <= '2020-05-31')
].copy()
```

```python
pollutants = ['PM2.5', 'PM10', 'NO', 'NO2', 'NOx', 'NH3', 'CO', 'SO2', 'O3', 'AQI']

pre_mean = pre_lockdown[pollutants].mean()
lock_mean = lockdown[pollutants].mean()

comparison = pd.DataFrame({
    'Pre-Lockdown (2019)': pre_mean,
    'Lockdown (2020)': lock_mean
})

comparison['Absolute Change'] = comparison['Lockdown (2020)'] - comparison['Pre-Lockdown (2019)']

comparison['% Change'] = np.where(
    comparison['Pre-Lockdown (2019)'] != 0,
    (comparison['Absolute Change'] / comparison['Pre-Lockdown (2019)']) * 100,
    np.nan
)

comparison = comparison.round(2).sort_values('% Change')

comparison
```

# Time Series Decomposition — AQI

## Objective

Decompose AQI time series into its fundamental components:

* **Trend** → long-term direction
* **Seasonality** → repeating patterns (e.g., winter spikes)
* **Residual (Error)** → random fluctuations

---

## Why Decomposition?

Air pollution data is not random. It contains:

* Strong seasonal effects
* Long-term trends
* Irregular noise

Decomposition helps isolate these components for better understanding and forecasting.

---


```python
from statsmodels.tsa.seasonal import seasonal_decompose

# Ensure datetime
delhi_data['Date'] = pd.to_datetime(delhi_data['Date'])

# Sort
delhi_data = delhi_data.sort_values('Date')

# Set index
df_ts = delhi_data.set_index('Date')

# Use AQI only
aqi_series = df_ts['AQI']

# Optional: resample monthly (recommended for clarity)
aqi_monthly = aqi_series.resample('ME').mean()
```

```python
result = seasonal_decompose(aqi_monthly, model='additive')
import matplotlib.pyplot as plt

fig = result.plot()
fig.set_size_inches(12,8)

plt.tight_layout()
plt.show()
```

# Yearly Quarterly AQI Comparison

## Objective

Analyze AQI trends across years by dividing data into quarters.

---

## Why Quarterly Analysis?

* Captures **seasonal variation**
* Enables **year-over-year comparison**
* Highlights **pollution patterns across seasons**

---

## Quarter Definition

| Quarter | Months    |
| ------- | --------- |
| Q1      | Jan – Mar |
| Q2      | Apr – Jun |
| Q3      | Jul – Sep |
| Q4      | Oct – Dec |

---

## Insight Goal

* Identify worst pollution quarters
* Compare AQI across years
* Detect seasonal patterns

---


```python
# Ensure datetime
delhi_data['Date'] = pd.to_datetime(delhi_data['Date'])

# Extract Year & Quarter
delhi_data['Year'] = delhi_data['Date'].dt.year
delhi_data['Quarter'] = delhi_data['Date'].dt.to_period('Q')
```

```python
# Extract quarter number
delhi_data['Q'] = delhi_data['Date'].dt.quarter

quarterly_table = delhi_data.groupby(['Year', 'Q'])['AQI'].mean().unstack()

quarterly_table.columns = ['Q1', 'Q2', 'Q3', 'Q4']

quarterly_table = quarterly_table.round(2)

quarterly_table
```
